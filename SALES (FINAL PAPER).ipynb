```python
import pandas as pd

# Load the dataset
file_path = '/Users/cylustariq/Desktop/sales_data.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Group the data by 'PRODUCTLINE' to identify the top-selling products by 'QUANTITYORDERED'
top_selling_products = data.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().sort_values(ascending=False)

# Print the top-selling products
print("Top-Selling Products:")
print(top_selling_products)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group the data by 'PRODUCTLINE' to identify the top-selling products by 'QUANTITYORDERED'
top_selling_products = data.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().sort_values(ascending=False)

# Visualize the top-selling products
plt.figure(figsize=(14, 7))
top_selling_products.plot(kind='bar', color='skyblue')
plt.title('Top Selling Products')
plt.ylabel('Quantity Ordered')
plt.xlabel('Product Line')
plt.xticks(rotation=45)
plt.show()
```

```python

# Calculate profit for each product as (SALES - (PRICEEACH * QUANTITYORDERED))
data['PROFIT'] = data['SALES'] - (data['PRICEEACH'] * data['QUANTITYORDERED'])
most_profitable_products = data.groupby('PRODUCTLINE')['PROFIT'].sum().sort_values(ascending=False)

# Print the most profitable products
print("Most Profitable Products:")
print(most_profitable_products)
```
