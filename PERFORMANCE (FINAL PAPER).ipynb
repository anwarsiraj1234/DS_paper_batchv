```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/Users/cylustariq/Downloads/Hybrid Exam Paper/Performance.csv'
data = pd.read_csv(file_path)
```

```python
# Descriptive statistics for numerical columns
descriptive_stats = data.describe()
print("Descriptive Statistics:\n", descriptive_stats)
```

```python
# Correlation analysis
correlation_matrix = data.corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python

# Grouped analysis by categorical variables
grouped_gender = data.groupby('gender').mean()
grouped_race = data.groupby('race/ethnicity').mean()
grouped_parent_education = data.groupby('parental level of education').mean()
grouped_lunch = data.groupby('lunch').mean()
grouped_test_prep = data.groupby('test preparation course').mean()
print("\nGrouped by Gender:\n", grouped_gender)
print("\nGrouped by Race/Ethnicity:\n", grouped_race)
print("\nGrouped by Parental Level of Education:\n", grouped_parent_education)
print("\nGrouped by Lunch:\n", grouped_lunch)
print("\nGrouped by Test Preparation Course:\n", grouped_test_prep)
```

```python
# Visualization setup
sns.set(style="whitegrid")
```

```python
# Histograms for scores
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(data['math score'], kde=True, color="skyblue", ax=axes[0], bins=20)
sns.histplot(data['reading score'], kde=True, color="olive", ax=axes[1], bins=20)
sns.histplot(data['writing score'], kde=True, color="gold", ax=axes[2], bins=20)
axes[0].set_title('Math Score Distribution')
axes[1].set_title('Reading Score Distribution')
axes[2].set_title('Writing Score Distribution')
# Show plots

plt.show()
```

```python
# Boxplots for scores by lunch type
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(x='lunch', y='math score', data=data, ax=axes[0])
sns.boxplot(x='lunch', y='reading score', data=data, ax=axes[1])
sns.boxplot(x='lunch', y='writing score', data=data, ax=axes[2])
axes[0].set_title('Math Scores by Lunch Type')
axes[1].set_title('Reading Scores by Lunch Type')
axes[2].set_title('Writing Scores by Lunch Type')
# Show plots
plt.show()
```

```python
# Print the descriptive statistics, correlations, and grouped means
print("Descriptive Statistics:\n", descriptive_stats)
print("\nCorrelation Matrix:\n", correlation_matrix)
print("\nGrouped by Gender:\n", grouped_gender)
print("\nGrouped by Race/Ethnicity:\n", grouped_race)
print("\nGrouped by Parental Level of Education:\n", grouped_parent_education)
print("\nGrouped by Lunch:\n", grouped_lunch)
print("\nGrouped by Test Preparation Course:\n", grouped_test_prep)
```
